```python
import pandas as pd 
import numpy as np
data=pd.read_csv('groceries - groceries.csv')
data
```

```python
data1=data.drop(['Item(s)'],axis=1)
data1
```

```python
data1=data1.replace(np.nan,'',regex=True)
data1
```

```python
data1
trans=[]
for i in range(0,9835):
    trans.append([str(data1.values[i,u])for u in range(1,31)])
trans
    
```

```python
gross_series=pd.DataFrame(pd.Series(trans))
gross_series
```

```python
gross_series.columns=['buisness']
gross_series
```

```python
x=gross_series['buisness'].str.join(sep=",").str.get_dummies(sep=",")
x
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
a=apriori(x,min_support=0.005,max_len=3,use_colnames=True)
```

```python
a
```

```python
a.sort_values('support',ascending=False,inplace=True)
a.sort_values
```

```python
rules=association_rules(a,metric='lift',min_threshold=1)
rules.sort_values('lift',ascending=False)
```
